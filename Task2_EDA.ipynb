```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
print("Libraries imported successfully!")
```

```python
data = {'Student':['Aman','Riya','Rahul','Neha','Karan','Priya','Arjun', 'Simran'],
'Age':[20,21,22,20,23,21,22,20],
'Marks':[78,85,67,90,72,88,75,95],
'Study_Hours':[4,6,3,7,5,6,4,8],
}
df = pd.DataFrame(data)

df
```

```python
print("Dataset Information:")
df.info()

print("\nMissing Values;")
print(df.isnull().sum())
```

```python
print("Basic Statistical Summary:")
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Student',y='Marks', data=df)

plt.title('Marks of Students')
plt.xlabel('Student Name')
plt.ylabel('Marks')

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Marks'], bins=5, edgecolor='black')

plt.title('Distribution og Student Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')

plt.show()
```

```python
numeric_df = df[['Age','Marks','Study_Hours']]

plt.figure(figsize=(7,5))
sns.heatmap(numeric_df.corr(),annot=True)

plt.title('Correlation Heatmap')
plt.show()
```

## Key insights
- Student marks vary across the dataset.
- Higher Study hours are generally associated with better marks.
- The bar chart compares the marks obtained by diffrent students.
- The hitorgram shows the distribution of students marks.
- The correlation heatmap shows the Ralatinship between Age, Marks, and Study hours.
## Conclusion 
The Exploratory Data Analysis helped in understanding the data set using statistical methods and visualizations. The analysis indicates the study hours have a positive relationship with student marks.
